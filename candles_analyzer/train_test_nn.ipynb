## Create dataset

```python
import logging
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import pandas_ta as ta

from preprocess import preprocess_data
```

```python
X = np.arange(2000).reshape(-1,1) / 50
Y = (np.sin(X) + 1) / 2
```

### Create model

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
import tensorflow.keras as keras

# Create model
# loss: 0.1842 - mae: 0.2779
# model = Sequential([
#     Dense(256, activation='tanh', input_shape=(1,)),
#     Dense(64, activation='tanh'),
#     Dense(32, activation='tanh'),
#     Dense(1, activation='tanh')
# ])

# Create model
model = Sequential([
    Dense(128, activation='sigmoid', input_shape=(1,)),
    Dense(32, activation='sigmoid'),
    Dense(16, activation='sigmoid'),
    Dense(1, activation='sigmoid')
])

# Compiling the neural network
model.compile(optimizer='adam',
              loss="mse",
              metrics="mae")

model.summary()

# Fitting the data to the training dataset
history = model.fit(X, Y, epochs=500, validation_data=(X, Y))
```

## Make test with prediction

```python
Y_pred = model.predict(X)

plt.figure(figsize=(20,10))
plt.plot(Y, color="green")
plt.plot(Y_pred, color="blue")
```
